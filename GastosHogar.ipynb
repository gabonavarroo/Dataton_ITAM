```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el shapefile de las entidades de México
mapa_mexico = gpd.read_file("gadm41_MEX_1.shp")

# Ejemplo de DataFrame con el número de entidad y gasto en salud
# Asegúrate de que el código de entidad coincida con el del shapefile
datos_salud = pd.DataFrame({
    'entidad': ['MEX.1_1'	, 'MEX.2_1', 'MEX.3_1'	, 'MEX.4_1'	, 'MEX.5_1', 	],  # Números de entidad
    'gasto_salud': [1000, 1500, 500, 2000, 1200]  # Gastos correspondientes
})

# Unir los datos de salud con el mapa usando el número de entidad
mapa_salud = mapa_mexico.merge(dfGastos, left_on="GID_1", right_on="entidadesM")

# Crear el mapa de calor
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
mapa_salud.plot(column='gasto', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Agregar título y etiquetas
ax.set_title("Mapa de calor: Gasto en salud por entidad en México", fontsize=15)
ax.set_axis_off()

plt.show()
```
