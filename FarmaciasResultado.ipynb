```python
import pandas as pd
import numpy as np
import os

df_comercio = pd.read_csv('denue_inegi_46321-46531_.csv', encoding = 'latin')
farmacias = df_comercio[df_comercio['nombre_act'].str.contains('farmacias', case = False, na = False)]
farm = farmacias

# Eliminamos las columnas que no nos son útiles por el momento
farmacias = farmacias.drop(columns = ['www','correoelec','telefono','tipo_v_e_1', 'nom_v_e_1',
       'tipo_v_e_2', 'nom_v_e_2', 'tipo_v_e_3', 'nom_v_e_3', 'numero_int', 'letra_int', 'tipoCenCom', 'nom_CenCom', 'numero_ext',
       'letra_ext', 'edificio', 'edificio_e', 'numero_int', 'letra_int', 'per_ocu', 'tipo_vial', 'tipoUniEco', 'id', 'clee', 'raz_social', 
       'nomb_asent', 'tipo_asent', 'nom_vial', 'fecha_alta', 'codigo_act', 'nombre_act', 'num_local'])
```

```python
#Contamos la cantidad de farmacias que hay por municipio
conteo_mun = farmacias['municipio'].value_counts().reset_index()
conteo_mun.columns = ['municipio', 'cont_fmun']
conteo_mun
```

```python
#Contamos farmacias por entidad federativa
conteo_estado = farmacias['entidad'].value_counts().reset_index()
conteo_estado.columns = ['entidad', 'contf_estado']
conteo_estado
```

```python
#Agregamos al DataFrame farmacias una columna donde diga cuantas farmacias tiene su municipio y estado y que se ordene de acuerdo al conteo por municipio
farmacias_conteo_mun = farmacias.merge(conteo_mun, on ='municipio', how = 'left')
farmacias_conteo_mun = farmacias_conteo_mun.merge(conteo_estado, on = 'entidad', how= 'left')
farmacias_ord_mun = farmacias_conteo_mun.sort_values(by = 'cont_fmun', ascending=True)
farmacias_ord_mun
```

```python
censo_habitantes = pd.read_csv('censo2020.csv', encoding = 'latin')
#habitantes_mun = censo_habitantes.drop(columns=['Porcentaje de hombres', 'Porcentaje de mujeres', 'Razón de dependencia', 'Relación \nhombres-mujeres', 'Porcentaje de la población estatal'])
habitantes_mun = censo_habitantes[~censo_habitantes['Municipio'].str.contains('000')]
habitantes_mun
```

```python
#Quitemos los numeros que van antes del municipio y guardemos nuestro nuevo string en una nueva columna llamada municipio
# En caso de que estas lineas fallen, volver a correr todo y correr esta nuevamente en particular
habitantes_mun['municipio'] = habitantes_mun['Municipio'].str.replace(r'^\d+\s+', '', regex=True)
habitantes_mun['entidad'] = habitantes_mun['entidad'].str.replace(r'^\d+\s+', '', regex=True)
habitantes_mun
```

```python
habitantes_mun['poblacion_total'] = habitantes_mun['poblacion_total'].str.replace(' ', '', regex=True)
habitantes_mun
```

```python
habitantes_mun = habitantes_mun[['entidad','municipio','poblacion_total','edad_mediana']]
habitantes_mun
```

```python
farmacias_ratio = farmacias_conteo_mun.merge(habitantes_mun, on = ['entidad','municipio'], how = "left")
farmacias_ratio
```

```python
#Realizaremos calculos para agregar una columna con la cantidad de farmacias por cada 1000 habitantes
farmacias_ratio['poblacion_total'] = pd.to_numeric(farmacias_ratio['poblacion_total'], errors= 'coerce')
farmacias_ratio['farm_por_hab'] = farmacias_ratio['cont_fmun'] / farmacias_ratio['poblacion_total'] * 1000
farmacias_ratio
```

```python
#Ordenamos de acuerdo a la cantidad de farmacias por cada 1000 personas
farmacias_ratio = farmacias_ratio.dropna(subset=['poblacion_total'])
farmacias_ratio_ord = farmacias_ratio.sort_values(by = 'farm_por_hab', ascending=True)
farmacias_ratio_ord
```

```python
farm_mun_pob = habitantes_mun
farm_mun_pob = farm_mun_pob.merge(conteo_mun, on = 'municipio', how = 'left')
farm_mun_pob = farm_mun_pob.merge(conteo_estado, on = 'entidad', how = 'left')
farm_mun_pob['cont_fmun'] = farm_mun_pob['cont_fmun'].fillna(0)
farm_mun_pob['contf_estado'] = farm_mun_pob['contf_estado'].fillna(0)

farm_mun_pob['poblacion_total'] = pd.to_numeric(farm_mun_pob['poblacion_total'], errors= 'coerce')
farm_mun_pob['farm_por_hab'] = farm_mun_pob['cont_fmun'] / farm_mun_pob['poblacion_total'] * 1000
farm_mun_pob['edad_mediana'] = habitantes_mun['edad_mediana']
```

```python
#Data frame de todos los municipios ordenado respecto a la cantidad de farmacias por cada 1000 habitantes
farm_mun_pob_ord = farm_mun_pob.sort_values(by= 'farm_por_hab', ascending = True)
farm_mun_pob_ord
```

```python
necesitadas_ratio = farm_mun_pob_ord[farm_mun_pob_ord['edad_mediana'] > 40]
necesitadas_ratio
```

```python
gastos_ingresos = pd.read_csv('GastosIngresosMunic.csv', encoding = 'latin')
gastos_ingresos = gastos_ingresos[['entidad', 'municipio', 'total_gastos', 'total_ingresos']]
gastos_ingresos.dropna()
gastos_ingresos
```

```python
data_mun = farm_mun_pob.merge(gastos_ingresos, on=['entidad','municipio'], how ='left')
data_mun
```

```python
data_mun['gastos_pc'] = data_mun['total_gastos'] / data_mun['poblacion_total']
data_mun['ingresos_pc'] = data_mun['total_ingresos'] / data_mun['poblacion_total']
data_mun
```

```python
#def filtrar_data(data, columna, umbral):
#   data = data.sort_value(by = columna, ascending = True)
 #   return data[data[columna] <= umbral]

umbral_ratio = 0.42
filtrado_ratio = data_mun.sort_values(by= 'farm_por_hab', ascending= True)
filtrado_ratio = filtrado_ratio[filtrado_ratio['farm_por_hab'] < umbral_ratio]
filtrado_ratio = filtrado_ratio.rename(columns= {'poblacion_total': 'poblacion_total_r', 'edad_mediana': 'edad_mediana_r',
                                                'cont_fmun':'cont_fmun_r', 'contf_estado':'contf_estado_r', 'farm_por_hab': 'farm_por_hab_r',
                                                 'total_gastos':'total_gastos_r', 'total_ingresos': 'total_ingresos_r',
                                                 'gastos_pc': 'gastos_pc_r', 'ingresos_pc': 'ingresos_pc_r'})
filtrado_ratio
```

```python
umbral_edad = 28
filtrado_edad = data_mun.sort_values(by= 'edad_mediana', ascending= False)
filtrado_edad = filtrado_edad[filtrado_edad['edad_mediana'] >= umbral_edad]
filtrado_edad = filtrado_edad.rename(columns= {'poblacion_total': 'poblacion_total_e', 'edad_mediana': 'edad_mediana_e',
                                                'cont_fmun':'cont_fmun_e', 'contf_estado':'contf_estado_e', 'farm_por_hab': 'farm_por_hab_e',
                                                 'total_gastos':'total_gastos_e', 'total_ingresos': 'total_ingresos_e',
                                                 'gastos_pc': 'gastos_pc_e', 'ingresos_pc': 'ingresos_pc_e'})
filtrado_edad
```

```python
umbral_gastos = 0.31
filtrado_gastos = data_mun.sort_values(by= 'gastos_pc', ascending= False)
filtrado_gastos = filtrado_gastos[filtrado_gastos['gastos_pc'] >= umbral_gastos
                                  ]
filtrado_gastos = filtrado_gastos.rename(columns= {'poblacion_total': 'poblacion_total_g', 'edad_mediana': 'edad_mediana_g',
                                                'cont_fmun':'cont_fmun_g', 'contf_estado':'contf_estado_g', 'farm_por_hab': 'farm_por_hab_g',
                                                 'total_gastos':'total_gastos_g', 'total_ingresos': 'total_ingresos_g',
                                                 'gastos_pc': 'gastos_pc_g', 'ingresos_pc': 'ingresos_pc_g'})
filtrado_gastos
```

```python
umbral_ingresos = .41
filtrado_ingresos = data_mun.sort_values(by= 'ingresos_pc', ascending= False)
filtrado_ingresos = filtrado_ingresos[filtrado_ingresos['ingresos_pc'] >= umbral_ingresos]
filtrado_ingresos= filtrado_ingresos.rename(columns= {'poblacion_total': 'poblacion_total_i', 'edad_mediana': 'edad_mediana_i',
                                                'cont_fmun':'cont_fmun_i', 'contf_estado':'contf_estado_i', 'farm_por_hab': 'farm_por_hab_i',
                                                 'total_gastos':'total_gastos_i', 'total_ingresos': 'total_ingresos_i',
                                                 'gastos_pc': 'gastos_pc_i', 'ingresos_pc': 'ingresos_pc_i'})
filtrado_ingresos
```

```python
ratio_edad = pd.merge(filtrado_ratio, filtrado_edad, on = ['entidad', 'municipio'], how = 'inner')
ratio_edad_gastos = pd.merge(ratio_edad, filtrado_gastos,on = ['entidad', 'municipio'], how = 'inner' )
farmacias_recomendadas = pd.merge(ratio_edad_gastos, filtrado_ingresos, on=['entidad', 'municipio'], how = 'inner' )
farmacias_recomendadas = farmacias_recomendadas[['entidad','municipio','poblacion_total_r','edad_mediana_r', 'cont_fmun_r', 'contf_estado_r', 'farm_por_hab_r','total_gastos_r','total_ingresos_r','gastos_pc_r','ingresos_pc_r']]
#filtrado_ratio = filtrado_ratio.rename(columns= {'poblacion_total_r': 'poblacion_total', 'edad_mediana_r': 'edad_mediana',
     #                                           'cont_fmun_r':'cont_fmun', 'contf_estado_r':'contf_estado', 'farm_por_hab_r': 'farm_por_hab',
      #                                           'total_gastos_r':'total_gastos', 'total_ingresos_r': 'total_ingresos',
       #                                          'gastos_pc_r': 'gastos_pc', 'ingresos_pc_r': 'ingresos_pc'})
farmacias_recomendadas
```
