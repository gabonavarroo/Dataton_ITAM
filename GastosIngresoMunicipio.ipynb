```python
import os
import pandas as pd

# Función para leer y procesar cada archivo CSV
def procesar_csv(ruta_archivo):
    # Leer el archivo CSV
    df = pd.read_csv(ruta_archivo, index_col=False)
    return df

# Obtener una lista de todos los archivos CSV en el directorio actual
archivos_csv = [archivo for archivo in os.listdir() if archivo.endswith('.csv')]

# Procesar cada archivo y almacenarlos en una lista
lista_dfs = []
for archivo in archivos_csv:
    df_procesado = procesar_csv(archivo)  # Procesar cada CSV
    lista_dfs.append(df_procesado)

# Concatenar todos los DataFrames en uno solo
df_concatenado = pd.concat(lista_dfs, ignore_index=True)

# Eliminar filas duplicadas
df_concatenado = df_concatenado.drop_duplicates()

# Mostrar los primeros datos procesados
df_concatenado.head()
```

```python
df_concatenado=df_concatenado[df_concatenado.MUNICIPIO!=' '][['ENTIDAD','MUNICIPIO','A700A','A800A','K000A','M000A']] ##IGNORA MUNICIPIOS EN BLANCO, DEJA COLUMNAS DE INTERES
```

```python
df_concatenado
```

```python
df_concatenado.dropna(subset=['A700A','A800A','K000A',	'M000A'])
```

```python
df_concatenado.MUNICIPIO=df_concatenado.MUNICIPIO.astype(int) ##CONVIERTO LA COLUMNA A INT PARA ELIMINAR LOS 0
```

```python

df_totMunici=df_concatenado.groupby(['ENTIDAD','MUNICIPIO']).sum() ##SUMA POR ENTIDAD Y MUNICIPIO
```

```python
df_totMunici.reset_index(inplace=True) #RESETEAR EL INDICE DE LAS FILAS
```
